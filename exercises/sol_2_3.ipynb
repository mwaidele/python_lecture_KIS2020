## 2. Using matplotlib objects implicitely vs. explicitely

Have a look at the code below. It uses an implicit way of handling axes and figures, to plot some physical behaviour. 
1. Without changing the outcome of the code, rewrite it such that objects of figures and axes are used explicitely.
2. Add a suiting title, labels to both axes, a legend depicting different `exp_fac`, add a grid and customize the y-axis ticks to only show the value 7, 9 and 11.

````python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt


def R(x, exp_fac=0):
    """Contraction R of some body as a function of force applied x, 
    and exp_fac."""
    return 10 - np.sqrt(2 * x) + np.exp(-exp_fac * x)


x = np.linspace(0, 2 * np.pi, 101)

plt.figure(1, figsize=(8, 6))
fig, ax = plt.subplots(1, 1, num=1)

exp_fac_list = [-0.1, 0, .1, 1.1, 10]
for elm in exp_fac_list:
    ax.plot(x, R(x, exp_fac=elm), label='exp_fac = %.1f' % elm)

ax.set_title('Contraction $R$')
ax.legend()
ax.set_xlabel('$x$'), ax.set_ylabel('$R$')
ax.grid()
ax.set_yticks([7, 9, 11])
````
